```python
from ultralytics import YOLO
import torch

DATA_PATH = "/content/data.yaml"
SAVE_DIR = "/content/drive/MyDrive/yolo_lung_segmentation_results"

print("GPU:", torch.cuda.get_device_name(0))

model = YOLO("yolo26m-seg.pt")

model.train(
    data=DATA_PATH,

    epochs=200,
    imgsz=1024,
    batch=6,
    device=0,

    optimizer="AdamW",
    lr0=3e-4,
    cos_lr=True,
    weight_decay=5e-4,
    scale=0.2,

    mask_ratio=2,

    patience=40,

    project=SAVE_DIR,
    name="train_lung_seg_yolo26m",
    exist_ok=True
)
```
