```python
from ultralytics import YOLO
import torch

DATA_PATH = "/content/data.yaml"
PROJECT_PATH = "/content/drive/MyDrive/yolo_lung_segmentation_results"

print("GPU:", torch.cuda.get_device_name(0))

model = YOLO("yolov8m-seg.pt")

model.train(
    data=DATA_PATH,
    task="segment",

    epochs=200,
    imgsz=1024,
    batch=6,
    device=0,

    optimizer="AdamW",
    lr0=3e-4,
    cos_lr=True,
    weight_decay=5e-4,
    scale=0.2,

    multi_scale=False,
    amp=True,
    workers=2,
    cache=False,

    mask_ratio=2,

    patience=40,

    project=PROJECT_PATH,
    name="train_lung_seg_optimized",
    exist_ok=True
)
```
